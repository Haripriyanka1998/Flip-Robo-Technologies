Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time

```python
driver=webdriver.Chrome("chromedriver.exe") 


url = "https://www.bcci.tv/"
driver.get(url)
```

```python
international_fixtures = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")
international_fixtures.click()
international_fixtures=driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']")
international_fixtures.click()
```

```python
# finding the element where match title is present
Match_title=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
Match_title
```

```python
Match_tags_list=[]
for i in Match_title:
    Match_tags_list.append(i.text)
Match_tags_list
```

```python
series=driver.find_elements_by_xpath("//div[@class='fixture__teams']")
series
```

```python
series_list=[]
for i in series:
    series_list.append(i.text)
series_list
```

```python
# finding the element where series is pres
place_element=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']")
place_element
```

```python
place_list=[]
for i in series:
    place_list.append(i.text)
place_list
```

```python
# finding the element where series is present
time_date=driver.find_elements_by_xpath("//div[@class='fixture__full-date']")
time_date
```

```python
time_date_list=[]
for i in time_date:
    time_date_list.append(i.text)
time_date_list
```

```python
print(len(Match_tags_list),len(series_list),len(place_list),len(time_date_list))
```

```python
import pandas as pd
ODI=pd.DataFrame({})
ODI['Match_tags_list']=Match_tags_list
ODI['series_list']=series_list
ODI['place_list']=place_list
ODI['time_date_list']=time_date_list
```
